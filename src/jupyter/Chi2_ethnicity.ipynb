```python
import pandas as pd
import numpy as np
import plotly.express as px
import pickle
import plotly.graph_objects as go
pd.options.plotting.backend = "plotly"
import scipy.stats as stats
from numpy import linspace
from numpy.core.umath import pi
```

```python
parent_df_path = 'usage_per_user/'

with open(parent_df_path+'joh_df.pkl', 'rb') as f:
    joh_per_usage = pickle.load(f)
    
with open(parent_df_path+'lon_df.pkl', 'rb') as f:
    lon_per_usage = pickle.load(f)
    
with open(parent_df_path+'nyc_df.pkl', 'rb') as f:
    nyc_per_usage = pickle.load(f)
    
with open(parent_df_path+'ran_df.pkl', 'rb') as f:
    ran_per_usage = pickle.load(f)


all_usage = pd.concat([joh_per_usage,lon_per_usage,nyc_per_usage,ran_per_usage],ignore_index = True)
```

```python
all_usage.drop(labels = all_usage[all_usage.ethnicity.eq('other')].index,inplace =True)
all_usage.reset_index(inplace = True,drop = True)
all_usage
```

```python
emoji_prop = all_usage.drop(columns = ['tweets_contain_emoji','total_tweets','gender','ethnicity'])
for index, row in emoji_prop.iterrows():
    s = emoji_prop.loc[index].sum()
    if s!=0:
        emoji_prop.loc[index] = emoji_prop.loc[index]/s
    else:
        emoji_prop.loc[index] = 0

emoji_prop_race = emoji_prop
emoji_prop_race['ethnicity'] = all_usage['ethnicity']
emoji_prop_race
```

```python
mean_prop_race = emoji_prop_race.mean()
mean_prop_asian = emoji_prop_race.loc[emoji_prop_race['ethnicity'] == 'asian'].mean()
mean_prop_black = emoji_prop_race.loc[emoji_prop_race['ethnicity'] == 'black'].mean()
mean_prop_hispanic = emoji_prop_race.loc[emoji_prop_race['ethnicity'] == 'hispanic'].mean()
mean_prop_white = emoji_prop_race.loc[emoji_prop_race['ethnicity'] == 'white'].mean()
```

```python
p_value_by_race = {}
for index,emoji in enumerate (mean_prop_race.index):
    
    X = emoji_prop_race[emoji]
    X = X.round(decimals=1)
    Y = emoji_prop_race['ethnicity']
    observed = pd.crosstab(Y,X) 
    
    chi2, p, dof, expected = stats.chi2_contingency(observed)
    p_value_by_race[emoji] = p 
```

```python
p_value_by_race_pd = pd.DataFrame(data=sorted(p_value_by_race.items(), key=lambda kv: kv[1]),columns=['Emoji','P-value'])
```

```python
for index,emoji in enumerate(p_value_by_race_pd['Emoji']):
    
    eth = ['asian','black','hispanic','white']
    idx = np.argmax([mean_prop_asian[emoji],mean_prop_black[emoji],mean_prop_hispanic[emoji],mean_prop_white[emoji]])
    p_value_by_race_pd .at[index,'ethnicity'] = eth[idx]
    
    
    
```
